```python
def impedanceToGamma(impedance, Z0 = 50):
    return (impedance - Z0)/(impedance + Z0)

def gammaToImpedance(gamma, Z0 = 50):
    """Gamma *(impedance + Z0) = (Impedance - Z0), hence
    Impedance*(1 - Gamma) = Gamma*Z0 + Z0, hence
    Imdepdance = Z0 * (Gamma + 1)/(1 - Gamma)"""
    return Z0*(gamma + 1.)/(1 - gamma)

def addConstantResistance(plot, Z0 = 50, colour="-k"):
    for r in [5,20,50,100,200]:
        Z = r*ones(500) + 1j*linspace(-500,500, num=500)
        gamma = impedanceToGamma(Z)
        if r == 50:
            plt.plot(angle(gamma), abs(gamma), "-b")
        else:
            plt.plot(angle(gamma), abs(gamma), colour)
        
def addConstantReactance(plot, Z0 = 50, colour="-r"):
    for XL in [-200,-100,-50,-20,5,0,5,20,50,100,200]:
        Z = linspace(0.1,500, num=500) + 1j*XL*ones(500)
        gamma = impedanceToGamma(Z)
        plt.plot(angle(gamma), abs(gamma), colour)
        
def addConstantConductance(plot, Z0 = 50, colour = "-m"):
    for g in [1./1500, 1./500, 1./200, 1./100, 1./50, 1./20, 1./5]:
        B = linspace(-1./5,1./5, num = 500)
        Y = g + 1j*B
        gamma = impedanceToGamma(1/Y)
        plt.plot(angle(gamma), abs(gamma), colour)
        
def transmissionLinePhase(l, f, lengthUnit = "m", frequencyUnit = "Hz", dielectric = "PTFE", deg = "True"):
    if frequencyUnit ==  "MHz":
        f = f*1e6
    elif frequencyUnit == "GHz":
        f = f*1e9
        
    if lengthUnit == "cm":
        l = l / 100.
    elif lengthUnit == "inch":
        l = l / 100. * 2.54

    if dielectric == "PTFE":
        epsilon_r = 2.1
        
    velocity = 1 / np.sqrt(epsilon_r)
        
    waveLength = velocity * c/f
    
    if deg:
        phase = l / waveLength * 360
    else:
        phase = l / waveLength * 2 * pi
        
    return phase


print "Phase of Cable at 285 MHz, 3\":", transmissionLinePhase(3, 285e6, lengthUnit = "inch")

print "Phase of Cable at 285 MHz, 6\":", transmissionLinePhase(6, 285e6, lengthUnit = "inch")

print "Phase of Cable at 71 MHz, 6\":", transmissionLinePhase(6, 71e6, lengthUnit = "inch")

print "Phase of Cable at 71 MHz, 24\":", transmissionLinePhase(24, 71e6, lengthUnit = "inch")


def addMatchingAndTuningRange(plot, nuc="13C", colour="y.", Xmin = 50, CTShunt = 0, CMShunt = 50e-12, label = "", number = 50, save = False):
    if nuc == "13C" or nuc == "1H":
        omega = gL.gammaList(nuc)*6.7
        print "omega: ", omega/(2*pi*1e6)
        XmaxT = 1./(omega*(1.5e-12 + CTShunt))
        XminT = 1./(omega*(45e-12 + CTShunt))

        XmaxM = 1./(omega*(1.5e-12 + CMShunt))
        XminM = 1./(omega*(45e-12 + CMShunt))

    
    reactanceRangeTune = linspace(XminT, XmaxT, num = number)
    reactanceRangeMatch = linspace(XminM, XmaxM, num = number)
    
    #print reactanceRange
    Z0 = 50
    
    Zlist = []
    for Xm in reactanceRangeMatch:
        for Xt in reactanceRangeTune:
            Z = Z0 + 1j*Xm
            Z = Z*(1j)*Xt/(Z + (1j)*Xt)
            Zlist.append(Z)
    
    Zlist = array(Zlist)
    #print Zlist
    gammaList = (Zlist - Z0)/(Zlist + Z0)
    #print gammaList
    plt.plot(angle(gammaList), abs(gammaList), colour, label = label)
    
    filteredList = []
    
    for g in gammaList:
        if abs(g) < 0.97:
            filteredList.append(g)
    
    impList = gammaToImpedance(np.array(filteredList))
        
    if save:
        #np.savetxt("matchTuneRange.txt", zip(angle(gammaList), abs(gammaList)), fmt = "%.4f")
        np.savetxt("matchTuneRange.txt", zip(np.real(impList)/50., np.imag(impList)/50.), fmt = "%.4f")

        
        
fig = plt.figure(figsize=(12,5), facecolor='white')
axes = plt.subplot(121, polar=True, axisbg='white')
#axes.spines['polar'].set_visible(False)
#plt.yticks([])
#plt.xticks([])


addConstantResistance(plt)
#addConstantReactance(plt, colour = "-gray")
addConstantConductance(plt)

addMatchingAndTuningRange(plt, Xmin=5, colour="c.", CTShunt = -0e-12, CMShunt = 0e-12)
#addMatchingAndTuningRange(plt, Xmin=5, colour="k.", CTShunt = 5e-12, CMShunt = 20e-12)
#addMatchingAndTuningRange(plt, Xmin=5, colour="b.", CTShunt = 15e-12, CMShunt = 20e-12)
#addMatchingAndTuningRange(plt, Xmin=5, colour="y.", CTShunt = 25e-12, CMShunt = 20e-12)
#addMatchingAndTuningRange(plt, Xmin=5, colour="m.", CTShunt = 35e-12, CMShunt = 20e-12)

addMatchingAndTuningRange(plt, Xmin=50, colour="r.")

plt.xlim([0,1])
#plt.grid(False)
#plt.polar?

f = 71.8e6
f = linspace(61.8e6, 81.8e6)
omega = 2*pi*f


Z0 = 50
L = 65e-9
# coil
coilImpedance = 0.5+1j*omega*L
gamma_coil = impedanceToGamma(coilImpedance)
#print "Gamma coil:", gamma_coil
plt.plot(angle(gamma_coil), abs(gamma_coil), "o")
plt.plot(0, 0, "ro")

#now add Shunt Capacitor
CT = 68e-12# + 15.2e-12
ZCT = -1j/(omega*CT)
Z = coilImpedance*ZCT/(coilImpedance+ZCT)
#print "Z after Shunt Capacitance", Z
gamma_shunt = impedanceToGamma(Z)
#print "Gamma: ", gamma_shunt
plt.plot(angle(gamma_shunt), abs(gamma_shunt), "om")

#now add Matching Capacitor
CM = 7.2e-12
ZCM = -1j/(omega*CM)
Z = Z + ZCM
#print "Z after Series Capacitance", Z
gamma_series = impedanceToGamma(Z)
#print "Gamma: ", gamma_series
plt.plot(angle(gamma_series), abs(gamma_series), "-.x")

#print "Resonance Frequency:", 1/sqrt(L*CT)/(2*pi)/1e6

plt.subplot(122)

plt.plot(f/1e6, abs(gamma_series))
```
