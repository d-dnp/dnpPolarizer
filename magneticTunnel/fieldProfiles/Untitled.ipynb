```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data500 = 1000*np.loadtxt("./fieldFile_2016715_1454_500_x_along_tunnel.txt")
data300 = 1000*np.loadtxt("./fieldFile_2016715_1459_300_x_along_tunnel.txt")



plt.plot(data500[:,0], label = "x")
plt.plot(data500[:,1], label = "y")
plt.plot(data500[:,2], label = "z")
plt.ylabel("Field (mT)")
plt.legend()

plt.figure()
plt.plot(data300[:,0], label = "x")
plt.plot(data300[:,1], label = "y")
plt.plot(data300[:,2], label = "z")
plt.ylabel("Field (mT)")
plt.legend()
```

```python
# for the other tunnel file that Javier send, z is along the long edge of the tunnel, and x is along the short edge of the tunnel.
# the magnetic field inside the tunnel then points in the direction (1, 0, 1) or e_x + e_z.
```
